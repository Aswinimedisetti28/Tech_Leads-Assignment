**4. list slicing**

#Introduction:

this is a part of python lists which the built-in methods ...

1. slice operation is used to print specific range of elements from the list.
2. slice operation is performed on lists with the use of colon(:).
3. L[:]- prints all the elements from the list .
4. L[2:]- prints all the elements from the list starting from index 2.
5. L[:5]- prints all the elements from the list till the ending index 4.
6. L[2:6]- prints the all elements from the list from the index 2 till index 5
7. L[-4:2]- prints the all elements from the list from the index -4 till index -3.
8. L[::2]-prints all the elements from the list with step 2[index+1element] from the beginning.
9. L[::-1]- prints all the elements from the list with step1[index+1] from the last.
10. L[1:6:3]-prints all the elements from the list from index 1 till index 5with step 3.

#py. programs of slicing lists

L=[10, 20, 30, 40, 50, 60]

print(L[:])
#output:[10, 20, 30, 40, 50, 60]

print(L[2:])
#output:[30,40,50,60]

print(L[:5])
#output:[10, 20, 30, 40, 50]

print(L[2:6])
#output:[20, 30, 40]

print(L[-4:-2])
#output:[30, 40]

print(L[::2])
#output:[10, 30, 50]

print(L[::-1])
#output:[60, 50, 40, 30, 20, 10]

print(L[1:6:3])
#output:[20, 30]
