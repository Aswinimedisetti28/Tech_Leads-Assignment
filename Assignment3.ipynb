#3.STRING FORMATTING I PYTHON:

Strings in python can be formatted with the use of format() method which is a very versatile and powerful tool for formatting strings . format method in string contains curly braces{} as placeholder which can hold arguments according  to position or keyword to specify the order.

**Output:**

**Python program for formatting of string:**

**Default order:**

X="{}, {}, {}".format('Hollywood', 'Bollywood', 'Kollywood')

print("Print string in default order:")

print(X)

**Output:**

Print string in default order:

Hollywood Bollywood Kollywood

**Positional formatting:**

X="{1}, {0}, {2}".format('Hollywood', 'Bollywood', 'Kollywood')

print("print string in positional order:")

print(X)

**Output**

print string in positional order:

Bollywood Hollywood Kollywood

**Keyword formatting**
X="{h}, {b}, {k}".format(h='Hollywood', b='Bollywood', k='Kollywood')

print("Print string in order of keyword:")

print(X)

**Output:**

print string in order of keyword:

Hollywood Bollywood Kollywood
