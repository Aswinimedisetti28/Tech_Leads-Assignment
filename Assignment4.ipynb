3.**Dictionary methods:**

python provides a variety of built- in methods that you can use to manipulate dictionaries. here are some commonly used dictionary methods:

1. **keys():**

Return a list of all the keys in the dictionary.

**Source code:**

food_list={"name":"chapati", "cost":30}

keys=food_list.keys()

**output:**

["name", "cost"]

2.**values():**

Returns a list of the values in the dictionary.

**Source code:**

my_dict = {"name": "John", "age": 30, "city": "New York"}

values = my_dict.values()

**Output:**

["John", 30, "New York"]

3.**items()**

Returns a list of tuples containing key-value pairs.

**Source code:**

my_dict = {"name": "John", "age": 30, "city": "New York"}

items = my_dict.items()

**Output**

[("name", "John"), ("age", 30), ("city", "New York")]

4.**get(key, default=None):**

Returns the value for the given key. if the key is not found , it returns the specified default value(or `None` by default).

5.**pop(key, default=None):**

Removes and returns the value for the given key. if the key is not found, it returns the specified default value.

6.**popitem():**

Removes and  returns a random key-value pair from the as a tuple.

7.**clear():**

Removes all key-values pairs from the dictionary, making it empty.

8.**update(other_dict):**

Updates the dictionary with key-values pairs from another dictionry or iterable.

9.**copy():**

Creates a shallow copy of the dictionary.

**Source code:**

original_dict = {"name": "John", "age": 30}

new_dict = original_dict.copy()

**Output:**

{"name": "John", "age": 30}

#Look into the difference between list and tuple:

Tuples and lists are both fundamental data structures in python , but they have some key difference in terms of mutability , usage, and vehaviour. Here's a comparison between tuples and lists:

**1.Mutability:**

**.Tuples:**

Tuples are immutable, meaning their elements cannot be changed or modified after creation . once a tuple is created, you cannot add, remove , or modify elements.

**.Lists:**

Lists are mutable , so you can change , add, or remove elements after the lists is created.



**2.Syntax:**

**.Tuples:**

Tuples are defined using parentheses`()` or without any delimiters. Elements are separated by commas.

**Lists:**

Lists are defined using square `[]`. elements are also separated by commas.

**3.Use cases:**

**.Tuples:**

Tuples are used when want to you want to store a collection of related items that shouldn't change.

**.Lists:**

Lists are used when you need a dynamic collection of elements that you might need to modify.

**4.Methods and operations:**

**.Tuples:**

Tuples have fewer builit-in methods compared to lists because of their immutability. they support basi operations like indexing and slicing.

**Lists:**

Lists have a wide range of built-in methods for adding, removing , and manipulating elements , making them more veratile.

**5.Iteration:**

**Tuples:**

Tuples can be iterated over using loops , just loops, just like  lists.

**Lists:**

Lists can be iterated over using loops, comprehensions and other iteration techniques.
